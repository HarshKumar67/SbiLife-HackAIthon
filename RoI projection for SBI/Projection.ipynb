```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the updated dataset
df = pd.read_csv("updated_propensity_score_dataset.csv")

# Set the style for seaborn
sns.set_style("whitegrid")

# Plot 1: ROI of the current policy
plt.figure(figsize=(10, 5))
sns.histplot(df['Policy_Maturity_Amount'] / df['Total_Premium_Paid'], bins=30, kde=True)
plt.axvline(1, color='r', linestyle='dashed', label="Break-even")
plt.title("Distribution of ROI for Purchased Policies")
plt.xlabel("ROI (Maturity Amount / Total Premium Paid)")
plt.ylabel("Number of Policies")
plt.legend()
plt.show()

# Plot 2: Alternative Market Growth vs. Policy ROI
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['Estimated_Alternative_Growth'], y=df['Policy_Maturity_Amount'], hue=df['Policy_Type'])
plt.plot([df['Estimated_Alternative_Growth'].min(), df['Estimated_Alternative_Growth'].max()],
         [df['Estimated_Alternative_Growth'].min(), df['Estimated_Alternative_Growth'].max()], 'r--', label="Equal Growth Line")
plt.xlabel("Alternative Market Growth")
plt.ylabel("Policy Maturity Amount")
plt.title("Market Trend Analysis: Insurance vs. Alternative Investments")
plt.legend()
plt.show()
```

## 1️⃣ Customer Age vs. Investment in Policy

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['Age'], y=df['Total_Premium_Paid'])
plt.title("Investment in Policy Across Different Age Groups")
plt.xlabel("Age")
plt.ylabel("Total Premium Paid")
plt.xticks(rotation=45)
plt.show()
```

## 2️⃣ Policy Type Distribution

```python
plt.figure(figsize=(8, 5))
df['Policy_Type'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Distribution of Different Policy Types")
plt.xlabel("Policy Type")
plt.ylabel("Number of Policies")
plt.show()
```

## 3️⃣ ROI Comparison: Policy vs Alternative Market Investments

```python
plt.figure(figsize=(10, 5))

# Ensure there are no division errors or infinite values
df['Policy_ROI'] = df['Policy_Maturity_Amount'] / df['Total_Premium_Paid']
df = df.replace([np.inf, -np.inf], np.nan).dropna(subset=['Policy_ROI', 'Estimated_Alternative_Growth'])

sns.kdeplot(df['Policy_ROI'], label="Policy ROI", fill=True)
sns.kdeplot(df['Estimated_Alternative_Growth'], label="Alternative Market Growth", fill=True)

plt.axvline(1, color='red', linestyle='dashed', label="Break-even")
plt.title("Comparison of Policy ROI vs Alternative Market Growth")
plt.xlabel("Return on Investment")
plt.ylabel("Density")
plt.legend()
plt.show()
```

## 4️⃣ Policy Hold Duration vs. Maturity Amount

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['Years_Completed'], y=df['Policy_Maturity_Amount'], hue=df['Policy_Type'])
plt.title("Policy Hold Duration vs. Maturity Amount")
plt.xlabel("Years Completed in Policy")
plt.ylabel("Maturity Amount")
plt.legend()
plt.show()
```

## 5️⃣ Premium Payment Trend Over the Years

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=df['Years_Completed'], y=df['Total_Premium_Paid'], marker='o')
plt.title("Total Premium Paid Over Time")
plt.xlabel("Years Completed in Policy")
plt.ylabel("Total Premium Paid")
plt.show()
```

## 6️⃣ Credit Score vs. Policy Investment

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['Credit_Score'], y=df['Total_Premium_Paid'])
plt.title("Impact of Credit Score on Investment in Policy")
plt.xlabel("Credit Score")
plt.ylabel("Total Premium Paid")
plt.show()
```

## 7️⃣ Engagement Score vs. Policy Purchase

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['Propensity_Score'], y=df['Total_Premium_Paid'])
plt.title("Customer Engagement vs. Policy Investment")
plt.xlabel("Engagement Score")
plt.ylabel("Total Premium Paid")
plt.show()
```

## 8️⃣ Policy Type vs. Expected Market Return

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['Policy_Type'], y=df['Estimated_Alternative_Growth'])
plt.title("Comparison of Policy Types with Market Growth")
plt.xlabel("Policy Type")
plt.ylabel("Estimated Alternative Market Growth")
plt.xticks(rotation=45)
plt.show()
```

## 9️⃣ Income vs. Insurance Investment Behavior

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['Annual_Income'], y=df['Total_Premium_Paid'], hue=df['Policy_Type'])
plt.title("Annual Income vs. Investment in Insurance")
plt.xlabel("Annual Income")
plt.ylabel("Total Premium Paid")
plt.show()
```

## 🔟 Risk vs. Return Analysis

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['Estimated_Alternative_Growth'], y=df['Policy_Maturity_Amount'], hue=df['Policy_Type'])
plt.axline((0, 0), slope=1, linestyle="--", color="red", label="Equal Growth")
plt.title("Risk vs. Return Analysis: Policies vs. Market")
plt.xlabel("Alternative Market Growth")
plt.ylabel("Policy Maturity Amount")
plt.legend()
plt.show()
```
